```python
from PIL import Image
import numpy as np

def read_hex_codes(file_path):
    with open(file_path, 'r') as file:
        hex_data = file.read().replace('\n', '').replace('0x', '').split(',')
    return hex_data

def hex_to_rgb565(hex_str):
    hex_val = int(hex_str, 16)
    r = (hex_val >> 11) & 0x1F
    g = (hex_val >> 5) & 0x3F
    b = hex_val & 0x1F
    # Convert to 8-bit per channel
    r = (r << 3) & 0xFF
    g = (g << 2) & 0xFF
    b = (b << 3) & 0xFF
    return r, g, b

def convert_to_image(hex_codes, width, height, bit_depth=32):
    if bit_depth == 16:
        image_array = np.zeros((height, width, 3), dtype=np.uint8)
    elif bit_depth == 32:
        image_array = np.zeros((height, width, 4), dtype=np.uint8)
    else:
        raise ValueError("Unsupported bit depth. Use 16 or 32.")
    for i in range(height):
        for j in range(width):
            index = i * width + j
            r, g, b = hex_to_rgb565(hex_codes[index])
            if bit_depth == 16:
                image_array[i, j] = [r, g, b]
            elif bit_depth == 32:
                image_array[i, j] = [r, g, b, 255]  # Alpha channel set to 255 (opaque)
    return image_array

def save_image(image_array, output_path, bit_depth=32):
    if bit_depth == 16:
        image = Image.fromarray(image_array, 'RGB')
    elif bit_depth == 32:
        image = Image.fromarray(image_array, 'RGBA')
    image.save(output_path)

def main():
    file_path = 'image_set.rtf'  # Change this to your actual file path
    output_image_path = 'output_image32.png'  # Output image file path

    width = 176  # Image width
    height = 144  # Image height
    bit_depth = 32  # Choose 16 or 32 for bit depth

    hex_codes = read_hex_codes(file_path)
    image_array = convert_to_image(hex_codes, width, height, bit_depth)
    save_image(image_array, output_image_path, bit_depth)
    print(f"Image saved to {output_image_path}")

if __name__ == "__main__":
    main()
```

```python
from PIL import Image
import binascii
import io

def read_hex_codes(file_path):
    with open(file_path, 'r') as file:
        hex_data = file.read().replace('\n', '')
    return hex_data

def hex_to_image(hex_string):
    # Convert hex string to binary data
    binary_data = binascii.unhexlify(hex_string)
    
    # Create an image from binary data
    image = Image.open(io.BytesIO(binary_data))
    
    return image

def save_image(image, output_path):
    image.save(output_path)

def main():
	file_path = 'captureimage.txt'
    output_image_path = 'output_imagetxt.png'  # Output image file path

    hex_string = read_hex_codes(file_path)
    image = hex_to_image(hex_string)
    save_image(image, output_image_path)
    print(f"Image saved to {output_image_path}")

if __name__ == "__main__":
    main()
```

```python
def read_hex_codes(file_path):
    with open(file_path, 'r') as file:
        hex_data = file.read().replace('\n', '').replace('0x', '').replace(',', '').replace(' ', '')
    return hex_data

def format_hex_string(hex_string, line_length=32):
    formatted_hex = ""
    for i in range(0, len(hex_string), line_length):
        formatted_hex += hex_string[i:i+line_length]
    return formatted_hex.strip()  # Remove the trailing newline

def main():
    # Assuming the hex codes are stored in a file called 'image_set.rtf'
    file_path = 'image_set.rtf'
    
    hex_string = read_hex_codes(file_path)
    formatted_hex_string = format_hex_string(hex_string)

    # Save the formatted hex string to a file (optional)
    output_file_path = 'formatted_hex_codes.txt'
    with open(output_file_path, 'w') as file:
        file.write(formatted_hex_string)
    
    print(f"Formatted hex string:\n{formatted_hex_string}")

if __name__ == "__main__":
    main()
```
